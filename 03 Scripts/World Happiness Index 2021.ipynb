```python
# List of columns with missing values
columns_with_missing = df_World.columns[df_World.isnull().any()]


# Impute numerical columns with mean
numerical_columns = df_World.select_dtypes(include=['float64']).columns
df_World[numerical_columns] = df_World[numerical_columns].fillna(df_World[numerical_columns].mean())

# Impute categorical columns with mode
categorical_columns = df_World.select_dtypes(include=['object']).columns
df_World[categorical_columns] = df_World[categorical_columns].fillna(df_World[categorical_columns].mode().iloc[0])

# Verify if all missing values are handled
missing_counts = df_World.isnull().sum()
print(missing_counts)
```

```python
# Check for duplicate rows
duplicate_rows = df_World[df_World.duplicated()]

# Print duplicate rows, if any
if not duplicate_rows.empty:
    print("Duplicate Rows:")
    print(duplicate_rows)
else:
    print("No duplicate rows found.")
```

```python
# List of columns with missing values
columns_with_missing = df_World2.columns[df_World2.isnull().any()]

# Impute numerical columns with mean
numerical_columns = df_World2.select_dtypes(include=['float64']).columns
df_World2[numerical_columns] = df_World2[numerical_columns].fillna(df_World2[numerical_columns].mean())

# Impute categorical columns with mode
categorical_columns = df_World2.select_dtypes(include=['object']).columns
df_World2[categorical_columns] = df_World2[categorical_columns].fillna(df_World2[categorical_columns].mode().iloc[0])

# Verify if all missing values are handled
missing_counts = df_World2.isnull().sum()
print(missing_counts)
```

```python
# Check for duplicate rows
duplicate_rows = df_World2[df_World2.duplicated()]

# Print duplicate rows, if any
if not duplicate_rows.empty:
    print("Duplicate Rows:")
    print(duplicate_rows)
else:
    print("No duplicate rows found.")
```

```python
# Check the range of years covered in df_World
print("Range of years covered in df_World:")
print("Min Year:", df_World['year'].min())
print("Max Year:", df_World['year'].max())

# Check if the 'year' column exists in df_World2
if 'year' in df_World2.columns:
    print("\nRange of years covered in df_World2:")
    print("Min Year:", df_World2['year'].min())
    print("Max Year:", df_World2['year'].max())
else:
    print("\n'year' column does not exist in df_World2.")

# Check if the data is up-to-date by comparing the maximum year with the current year
import datetime
current_year = datetime.datetime.now().year
if df_World['year'].max() == current_year:
    print("\nData in df_World is up-to-date.")
else:
    print("\nData in df_World is not up-to-date. Current year:", current_year)


```

```python
# Select columns with mixed data types
mixed_type_columns_world = df_World.select_dtypes(include=['object']).columns
mixed_type_columns_world2 = df_World2.select_dtypes(include=['object']).columns

# Print the mixed-type columns (if any)
if len(mixed_type_columns_world) > 0:
    print("Mixed-type columns in df_World:")
    print(mixed_type_columns_world)
else:
    print("No mixed-type columns found in df_World.")

if len(mixed_type_columns_world2) > 0:
    print("\nMixed-type columns in df_World2:")
    print(mixed_type_columns_world2)
else:
    print("No mixed-type columns found in df_World2.")
```

```python
# Convert 'Country name' column to string data type in df_World
df_World['Country name'] = df_World['Country name'].astype(str)

# Convert 'Country name' and 'Regional indicator' columns to string data type in df_World2
df_World2['Country name'] = df_World2['Country name'].astype(str)
df_World2['Regional indicator'] = df_World2['Regional indicator'].astype(str)
```
